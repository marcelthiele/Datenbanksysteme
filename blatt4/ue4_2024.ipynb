### 1) Initialisierung

```python
import sqlite3

pathToBase = "terraBase.db"
conn = sqlite3.connect(pathToBase)
cur = conn.cursor()


def executeQueryAndPrintResults(query):
    cur.execute(query)
    results = cur.fetchall()
    print(f"--> {len(results)} rows")
    for res in results:
        print(res)
    return
```

<span style="font-size:0.8em;">\[*Create the Ticket table. Use foreign keys to ensure that there is a tuple in the Fan relation and a concert in the Concert relation for each ticket.
```python
#Erstellen Sie die Tabelle `Ticket`. Stellen Sie mittels Fremdschlüsseln sicher, dass es zu jedem Ticket ein Tupel in der `Fan` Relation sowie ein Tupel in der `Concert` Relation gibt. Deklarieren Sie `concert_id`, `fan_id` und `purchase_time` zusammen als Primärschlüssel. Die restlichen Attribute sollen `NOT NULL` sein. Stellen Sie außerdem sicher, dass die Anzahl (`quantity`) der Tickets pro Kauf zwischen 1 und 10 liegt.
query = "CREATE TABLE Ticket (concert_id INTEGER NOT NULL, fan_id TEXT NOT NULL, purchase_time TIMESTAMP NOT NULL, quantity INTEGER NOT NULL CHECK(quantity BETWEEN 1 AND 10), PRIMARY KEY(concert_id, fan_id, purchase_time), FOREIGN KEY(concert_id) REFERENCES Concert(c_id), FOREIGN KEY(fan_id) REFERENCES Fan(name));"
executeQueryAndPrintResults(query)
```
